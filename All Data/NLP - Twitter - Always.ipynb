# Application of Deep Neural Network for Twitter Data

```python
#For this project we will only use panda and numpy
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
#This is one time thing, where we will download that data from online
df = pd.read_csv("C:/Users/pnkum/Desktop/Projects/NLP/Praveen/Always_Concrete.csv", encoding="latin-1")
df.shape
```

```python
df.info()
```

```python
df["class"].value_counts()
```

```python
features = list(df.columns[1:1228])
```

```python
target = df.columns[1228:1229]
```

```python
#Now we need to separate the data into feature data and target data
X = df[features] #our features that we will use to predict Y
Y = df[target] #our dependent variable, the one we are trying to predict from X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
config = tf.contrib.learn.RunConfig(tf_random_seed=42) # not shown in the config

feature_cols = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300,100], n_classes =20,
                                         feature_columns=feature_cols, config=config)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf) # if TensorFlow >= 1.1
dnn_clf.fit(X_train, y_train, batch_size=50, steps=40000)
```

```python
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

y_pred = dnn_clf.predict(X_test)
accuracy_score(y_test, y_pred['classes'])
```

```python
recall_score(y_test, y_pred['classes'], average='micro')
```

```python
precision_score(y_test, y_pred['classes'], average='micro')
```

```python
cohen_kappa_score(y_test, y_pred['classes'])
```

```python
precision_recall_fscore_support(y_test, y_pred['classes'])
```
